```python
# import the TrendReq method from the pytrends request module
from pytrends.request import TrendReq

# execute the TrendReq method by passing the host language (hl) and timezone (tz) parameters
pytrends = TrendReq(hl='pt-BR', tz=180)

kw_list = ["PETR4"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='BR', gprop='')

gt_o3_ytd1 = pytrends.interest_over_time()
gt_o3_ytd1
```

```python
import yfinance as yf

df = yf.download('PETR4.SA', period='5y')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize=(15,6), sharex=True)

# Trends
gt_o3_ytd1.PETR4.plot(color='blue')

# Adjusted Closing Prices
df['Adj Close'].shift(1).plot(color='orange')

plt.legend(['Trend', 'Adj Close'])

plt.show();
```
